```python
import pandas as pd
```

```python
df = pd.read_csv("/gdrive/MyDrive/datas.csv")
print(df.to_string())
```

```python
sutunlar = ["averageRating" , "releaseYear"]
yenidf = df[sutunlar]
print(yenidf.to_string())
```

```python
import matplotlib.pyplot as plt
bestMovies = df.loc[df.groupby('releaseYear')['averageRating'].idxmax()]
yayinlanma_yili = bestMovies['releaseYear']
puan = bestMovies['averageRating']
plt.plot(yayinlanma_yili, puan, marker='o', linestyle='-')
plt.xlabel('Yayınlanan Yıl')
plt.ylabel('Puan')
plt.title('Her Yılın En Yüksek Puanlı Filmi')
plt.grid(True)
plt.show()
```
